```python
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error
def evaluate(model,X_train,X_test,y_train,y_test):
    train_pred =model.predict(X_train)
    test_pred =model.predict(X_test)
    print(f"Accuracy for your {model} is:")
    print(f"accuracy_acore for training {model} is {accuracy_score(y_train,train_pred)}")
    print(f"accuracy_score for testing {model} is {accuracy_score(y_test,test_pred)}")
```

```python
# from sklearn.linear_model import LogisticRegression
# leg = LogisticRegression()
# leg.fit(X_train,y_train)
from sklearn.linear_model import LogisticRegression

leg = LogisticRegression(class_weight={0:1, 1:2})
leg.fit(X_train, y_train)
```

```python
evaluate(leg,X_train,X_test,y_train,y_test)
```

```python
y_prob = leg.predict_proba(X_test)[:,1]
y_pred = (y_prob > 0.4).astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier()
```

```python
rfc_param={
    'n_estimators':[100,200,300],
    'max_depth':[5,10,15],
    'min_samples_split':[5,10,15],
    'min_samples_leaf':[5,10,15],
    'max_features':['sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

rfc_search = GridSearchCV(
    estimator=rfc,
    param_grid=rfc_param,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_search.fit(X_train,y_train)
```

```python
best_model = rfc_search.best_estimator_
```

```python
evaluate(best_model,X_train,X_test,y_train,y_test)
```
